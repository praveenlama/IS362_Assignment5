#### What is the easternmost airport in the United States?


```python
eastern = pd.read_csv('https://raw.githubusercontent.com/maddogdatascience/DATA607/master/airports.csv')
eastern_airport = eastern[(eastern['lat'] > 44) & (eastern['lat'] <= 50) & (eastern['lon'] >= -75) & (eastern['lon'] <= -65)]
eastern_airport.head()
eastern_airport.sort_values('lon', ascending=False).head(5)
```

> Since the latitude and longitudinal values seemed accurate, I double-checked each value using www.airnav.com and www.latitude.to Each value matched the values in the table above. **Eastport Municipal Airport** is the earthernmost airport in the United States.

#### On February 12th, 2013, which New York area airport had the windiest weather?


```python
windy = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
# sorting data based on requirements
windy_weather = windy[(windy["month"]==2) & (windy["day"]==12) & (windy["year"]==2013)]
windy_weather.head()
windy_weather.sort_values("wind_speed", ascending=False).head(5)
```

> Since, LGA and JFK is the airport of NYC, **LGA** had the most windiest weather on Feb 12th , 2013.
